```python
#1.Give column name as Id, Movie, Year, Rating, Duration
import pandas as pd
df=pd.read_csv('/content/movies_cleaned.csv',
               names=['Id','Movie','Year','Rating','Duration'])
df
```

```python
#2.Display the number of rows and number of columns
df.shape
```

```python
#3.Display the five observations that have highest duration

df.sort_values(by='Duration').tail(5)
```

```python
#4.Remove if any duplicate exists and change the index accordingly
df.drop(['Id'],axis=1,inplace=True)
print(df)


```

```python
df.drop_duplicates(inplace=True)
```

```python
#5.Rename the column name ‘Movie’ as Film
df.rename(columns={'Movie': 'Film'},inplace=True)


```

```python
#6.Display all the observation starting from index 100 upto 200

df.iloc[100:200]

```

```python
#7.Display the name and releasing year of the film with maximum duration
df.sort_values('Duration',ascending=False).head(1)[['Film','Year']]
```

```python
#8.Display the count of films released in each year.

df['Year'].value_counts()
```

```python
#9.Find the film released between 1986-1996 having highest durdf[(df['Year'] >= 1986) & (df['Year'] <= 1996)].sort_values(by='Duration', ascending=False).head(1)
df.loc[(df['Year']>=1986)&(df['Year']<=1996)].sort_values(by='Duration',ascending=False).head(1)
```

```python
#10.Find the films having duration greater than 7000 and rating above 3.5
df.loc[(df['Duration'] > 7000) & (df['Rating'] > 3.5)]
```

```python
#11.Print the details about 5 films having less rating , display in the descending order
df.sort_values(by='Rating', ascending=False).tail(5)

```

```python
#12.Print the name of the film released in 1996 have highest rating
df.loc[df['Year'] == 1996].sort_values(by='Rating').tail(1)[['Film']]
```
