```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Datasets/Real estate.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.drop(['X1 transaction date','No'],axis=1, inplace=True)
df
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import StandardScaler
Scaler=StandardScaler()
X_Scaled=Scaler.fit_transform(X)
X_Scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_Scaled,y,test_size=0.3,random_state=18)
```

```python
from sklearn.linear_model import LinearRegression
mlr=LinearRegression()
mlr.fit(X_train,y_train)
```

```python
y_pred=mlr.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
print("r2_score",r2_score(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
y_predknn=knn.predict(X_test)
y_predknn
```

```python
X_test
```

```python
print("r2_score",r2_score(y_test,y_predknn))
```
