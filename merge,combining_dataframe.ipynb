```python
import pandas as pd
df1=pd.DataFrame({'name':['manu','krishna','anitha'],
                  'rollno':[11,45,23,],
                  'class':['BCA','BCA','BBA']})
df1
```

```python
df2=pd.DataFrame({'name':['anand','vinu','nitha'],
                  'rollno':[12,46,23],
                  'class':['BCA','BCA','BBA']})
df2
```

```python
df3=pd.DataFrame({'mark':[34,45,35],
                 'gender':['M','F','F']})
df3
```

```python
#df1,df2
df4=pd.concat([df1,df2],ignore_index=True)
df4
```

```python
#df1,pd3
df5=pd.concat([df1,df3],axis=1)
df5
```

```python
df6=pd.DataFrame({'rollno':[11,45,22,47,24,56],
                  'marks':[34,45,50,23,36,49]})
df6
```

```python
pd.merge(df4,df6,how='inner',on='rollno')
```

```python
pd.merge(df4,df6,how='left',on='rollno')
```

```python
pd.merge(df4,df6,how='right',on='rollno')
```
